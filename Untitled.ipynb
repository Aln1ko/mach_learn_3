Замінимо категоріальну зміну стать на числову

```python
sex_mapping = {'male': 1, 'female': 0}
train['Sex'] = train['Sex'].map(sex_mapping)
test['Sex'] = test['Sex'].map(sex_mapping)
test.head()
```

```python
train.head()
```

Заповнимо вік медіаною

```python
median_age = train['Age'].median()
train['Age'] = train['Age'].fillna(median_age)
test['Age'] = test['Age'].fillna(median_age)
print(train.isnull().sum().to_string())
```

```python
print(test.isnull().sum().to_string())
```

Заповнимо місце звідки виїхали модою

```python
mode_embarked = train['Embarked'].mode()[0]
train['Embarked'] = train['Embarked'].fillna(mode_embarked)
print(train.isnull().sum().to_string())
```

Заповнимо Fare медіаною

```python
median_Fare = test['Fare'].median()
test['Fare'] = test['Fare'].fillna(median_Fare)
print(test.isnull().sum().to_string())
```

Перетворимо фічу Embarked з категоріальною на 3 числові

```python
train = pd.get_dummies(train, columns=['Embarked'], dtype=int)
test = pd.get_dummies(test, columns=['Embarked'], dtype=int)
train.head()
```

За допомогою методу локтя знайдемо оптимальну кількість кластерів

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage

def optimal_clusters(data):
    data_dist = pdist(data, 'euclidean')
    data_linkage = linkage(data_dist, method='average')
    last = data_linkage[-10:, 2]
    acceleration = np.diff(last, 2)
    k = acceleration.argmax() + 2
    return k

numerical_features = train.select_dtypes(include = ['float64', 'int64']).columns
train_data = train[numerical_features]

numerical_features = test.select_dtypes(include = ['float64', 'int64']).columns
test_data = test[numerical_features]

scaler = StandardScaler()
X = scaler.fit_transform(train_data)
Y = scaler.fit_transform(test_data)

iters = []
for k in range(1, 20):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    iters.append(kmeans.inertia_)

plt.plot(range(1, 20), iters, marker='o')
plt.title('Elbow_M')
plt.xlabel('N_clusters')
plt.ylabel('iters')
plt.show()

recommended_clusters_train = optimal_clusters(X)
print("Recommended number of clusters for train data:", recommended_clusters_train)
```

Проведемо навчання та намалюємо кластери для рекомендованого значення

```python
from sklearn.decomposition import PCA

cluster = KMeans(n_clusters=9, random_state=42)
labels = cluster.fit_predict(X)

pca = PCA(n_components = 2)
principal_components = pca.fit_transform(X)

plt.scatter(principal_components[:, 0], principal_components[:, 1], c=labels, cmap='rainbow')
plt.title('Clustered')
plt.show()

cluster = KMeans(n_clusters=9, random_state=42)
labels = cluster.fit_predict(Y)
pca = PCA(n_components = 2)
principal_components = pca.fit_transform(Y)

plt.scatter(principal_components[:, 0], principal_components[:, 1], c=labels, cmap='rainbow')
plt.title('Clustered')
plt.show()

```
